```python
for column_name in features.columns:
    print(column_name)
    x = features[column_name]
    norm_counts = pd.DataFrame(
        {"count": x.value_counts(), "prop": x.value_counts(normalize=True)}
    )
    display(norm_counts)
    if len(norm_counts) <= 24:
        sns.countplot(x=x)
#         plt.text(v, i, " "+str(v), color='blue', va='center', fontweight='bold')
        plt.xticks(rotation="vertical")
        plt.show()
```
